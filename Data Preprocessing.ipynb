Importing necessary libraries


```python
import pandas as pd
import json 
```

Match Result :

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'}, axis = 1, inplace = True)
```

```python
df_match.head()
```

Creating a dict to join the match_id with the Batting DataSet:

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
match_ids_dict

```

```python
match_ids_dict["Namibia VS Sri Lanka"]
```

Batting Summary :

```python
with open ('t20_json_files/t20_wc_batting_summary.json') as f :
    data =json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

Changing The Dismissal Column Name To Out/Notout :

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head(10)
```

Dropping The Dismissal Column :

```python
df_batting.drop(columns=["dismissal"],inplace = True )
df_batting.head()
```

```python
df_batting.head(5)
```

Removing Special Charaters from the Data:

```python
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x : x.replace('â€(c)\xa0',''))
```

```python
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x : x.replace('(c)\xa0',''))
```

```python
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x : x.replace('(c)',''))
```

```python
df_batting.head(5)
```

```python
match_ids_dict["Pakistan VS Netherlands"]
```

Joining match_ids_dict from Match Dataset To Batting Dataset and creating a new column in batting dataset as match_id

```python
df_batting.drop(columns=["match_id","mach_id"],inplace = True )
df_batting.head()
```

df_batting["mach_id"] = df_batting["match"].map(match_ids_dict)

df_batting.head(50)

```python
df_batting.shape
```

Exporting Data To CSV : 

```python
df_batting.to_csv('t20_csv_files/fact_bating_summary.csv', index = False)
```

<h4> (3) Process Bowling Summary</h4>

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```
